## Split the data into train and test and calculate the accuracy of the prediction

```python
# Function to sort eigenvalues and eigenvectors
def eigsort(A):
    eigvals, U = np.linalg.eig(A)
    order = np.argsort(eigvals)[::-1]
    eigvals = eigvals[order]
    U = U[:, order]
    return U, eigvals

# Function to perform classic PCA
def pca_classic(X, M):
    #για κάθε στήλη του πίνακα X υπολογίστε το μέσο όρο, και αποθηκεύστε το σε διάνυσμα γραμμής
    mu = X.mean(axis=0).reshape( (1,-1) )
    #αφαιρέστε το μέσο όρο από τον πίνακα
    normalized_X = X-mu
    #υπολογίστε το Covariance matrix για τον πίνακα του προηγούμενου βήματος
    S = (1/X.shape[0]) * normalized_X.T.dot( normalized_X )
    #υπολογίστε τα ιδιοδιανύσματα και τις ιδιοτιμές του covariance πίνακα
    eigvectors, eigvals = eigsort( S )
    #κρατήστε στον πίνακα U τις πρώτες Μ στήλες των ιδιοδιανυσμάτων
    U = eigvectors[:,:M]
    #κρατήστε στον πίνακα Lambads τις πρώτες Μ ιδιοτιμές
    Lambdas = eigvals[:M]
    print( X.shape, U.shape, Lambdas.shape )
    #αποθηκεύστε το μετασχηματισμό Z = U * X ( προσέξτε τις διαστάσεις )
    #ο πίνακας U εφαρμόζεται στον κανονικοποιημένο πίνακα X και τον προβάλει
    #επάνω στα ιδιοδιανύσματα(μειωμένη διάσταση)
    Z = normalized_X.dot(U)
    return Z, U, Lambdas, mu

# Standardize the iris dataset

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_iris)

# Apply PCA to the Iris dataset (reducing to 2 dimensions)
M = 2
Z, eigvecs, eigvals, mu = pca_classic(X_scaled, M)
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(Z, y_iris, test_size=0.3, random_state=42)

# Train a Logistic Regression classifier
clf = LogisticRegression(random_state=42)
clf.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Classification accuracy: {accuracy:.2f}")
```

## Introduction
This notebook demonstrates the implementation and application of the least squares algorithm,
traditionally used for regression, adapted for a classification task on the Iris dataset.

## Overview of Iris Dataset
- **Dataset**: Iris Flower Dataset
- **Classes**: Setosa, Versicolor, Virginica (3 classes)
- **Features**: Sepal Length, Sepal Width, Petal Length, Petal Width (4 features)


## Data Loading and Preprocessing

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Feature Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## Adapting Least Squares for Classification

The least squares method, typically used for regression, can be adapted for classification tasks.
For a multiclass classification like the Iris dataset, we use the one-vs-rest approach.
This involves training a separate linear model for each class.


```python
def train_least_squares(X, y, class_label):
    """
    Trains a least squares classifier for a given class.
    """
    # Create a binary target vector for the current class
    y_binary = np.where(y == class_label, 1, -1)

    # Add a bias term (1s) to the feature matrix
    X_bias = np.hstack([np.ones((X.shape[0], 1)), X])

    # Calculate the weights using the normal equation
    weights = np.linalg.pinv(X_bias.T @ X_bias) @ X_bias.T @ y_binary
    return weights

# Training a model for each class
weights_setosa = train_least_squares(X_train, y_train, 0)
weights_versicolor = train_least_squares(X_train, y_train, 1)
weights_virginica = train_least_squares(X_train, y_train, 2)

def predict(X, weights_setosa, weights_versicolor, weights_virginica):
    """
    Makes predictions using the trained models.
    """
    # Add a bias term to the input features
    X_bias = np.hstack([np.ones((X.shape[0], 1)), X])

    # Calculate the scores for each class
    scores_setosa = X_bias @ weights_setosa
    scores_versicolor = X_bias @ weights_versicolor
    scores_virginica = X_bias @ weights_virginica

    # Stack scores and predict the class with the highest score
    scores_all = np.vstack([scores_setosa, scores_versicolor, scores_virginica]).T
    predictions = np.argmax(scores_all, axis=1)
    return predictions

# Making predictions on the test set
y_pred = predict(X_test, weights_setosa, weights_versicolor, weights_virginica)
```

**Implementation of OvR with Least Squares**

Training Separate Models:

The function train_least_squares(X, y, class_label) is defined to train a binary classifier using the least squares method for each class.
For each class (Setosa, Versicolor, Virginica), it creates a binary target vector y_binary where the current class is labeled as 1 (positive class), and all other classes are labeled as -1 (negative class).
The feature matrix X is augmented with a bias term to include the intercept in the linear model.
Weights Calculation:

The least squares solution is computed using the normal equation, which involves calculating the pseudoinverse of the design matrix (X_bias) and multiplying it with the binary target vector y_binary. This results in a set of weights for each class.
Training for Each Class:

Separate models (weights) are trained for each class:
weights_setosa for Setosa vs. Not Setosa
weights_versicolor for Versicolor vs. Not Versicolor
weights_virginica for Virginica vs. Not Virginica
Prediction and Decision Rule
Prediction Function:

The predict function is used for making predictions. It takes the feature matrix X and the weights of each classifier as inputs.
Calculating Scores:

For a given input sample, the function calculates a score for each class by multiplying the input features (augmented with a bias term) with the weights of each classifier.
Selecting the Class:

The scores across all classifiers are combined, and the class corresponding to the highest score is chosen as the prediction for each sample.
Analysis of What OvR Does in This Context
Binary Classification for Multiclass Problem:

OvR simplifies the multiclass classification problem into multiple binary classification problems. Each binary classifier focuses on distinguishing one class from all others.

Model Training:
Separate models are trained for each class, allowing the algorithm to learn specific features that are unique to each class.

Decision Making:
During prediction, the approach leverages the strength of each binary classifier to identify the most likely class. The class with the highest score from its respective classifier is chosen, reflecting the classifier's confidence that the sample belongs to that class.

Adaptation of Least Squares:
The least squares method, typically used for regression, is effectively adapted for classification. While it doesn't inherently provide probabilistic outputs like logistic regression, the highest score among classifiers is used as a proxy for class selection.
In summary, the OvR approach in this implementation effectively adapts the least squares method for a multiclass classification problem. It trains individual models for each class and uses these models collectively to predict the class of new samples based on which model gives the highest score.

## Implementation of Naïve Bayes Classifier

The Naïve Bayes algorithm is a probabilistic classifier that applies Bayes' theorem with the 'naïve' assumption of conditional independence among features.
It's particularly suited for high-dimensional datasets.


```python
class NaiveBayesClassifier:
    def fit(self, X, y):
        # Determine the number of samples and features in the dataset
        n_samples, n_features = X.shape

        # Find the unique classes in the target variable
        self.classes = np.unique(y)
        n_classes = len(self.classes)

        # Initialize arrays for the mean, variance, and prior probabilities of each class
        self.mean = np.zeros((n_classes, n_features), dtype=np.float64)
        self.var = np.zeros((n_classes, n_features), dtype=np.float64)
        self.priors = np.zeros(n_classes, dtype=np.float64)

        # Calculate mean, variance, and prior probability for each class
        for c in self.classes:
            X_c = X[y == c]
            self.mean[c, :] = X_c.mean(axis=0)
            self.var[c, :] = X_c.var(axis=0)
            self.priors[c] = X_c.shape[0] / float(n_samples)

    def predict(self, X):
        # Predict the class for each sample in X
        y_pred = [self._predict(x) for x in X]
        return np.array(y_pred)

    def _predict(self, x):
        # Calculate the posterior probability for each class
        posteriors = []

        for idx, c in enumerate(self.classes):
            # Calculate the log of prior probability
            prior = np.log(self.priors[idx])
            # Sum the log of the class conditional probabilities
            class_conditional = np.sum(np.log(self._pdf(idx, x)))
            # Calculate posterior probability
            posterior = prior + class_conditional
            posteriors.append(posterior)

        # Return the class with the highest posterior probability
        return self.classes[np.argmax(posteriors)]

    def _pdf(self, class_idx, x):
        # Calculate the Gaussian probability density function for a given class and feature
        mean = self.mean[class_idx]
        var = self.var[class_idx]
        numerator = np.exp(- (x-mean)**2 / (2 * var))
        denominator = np.sqrt(2 * np.pi * var)
        return numerator / denominator

```

## Model Training and Prediction

The model is trained on the training dataset and then used to predict the labels of the test dataset.


```python
# Create and train the model
model = NaiveBayesClassifier()
model.fit(X_train, y_train)

# Predict labels for the test set
y_pred = model.predict(X_test)
```
